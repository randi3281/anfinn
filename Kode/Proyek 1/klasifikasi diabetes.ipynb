```python
import pandas as pd

data = pd.read_csv('..\..\Dataset\diabetes.csv', sep=',')
data.head()

# Outcome 1 artinya positif diabetes, 0 artinya negatif diabetes
```

```python
data_feature = data.drop(columns='Outcome')
data_label = data['Outcome']

print(data_feature.head())
print(" ")
print(data_label.head())
```

```python
x = data_feature.to_numpy()
y = data_label.to_numpy().reshape(-1, 1)
print(len(x))
print(len(y))
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import ModelCheckpoint

# Define the model
model = Sequential([
    Dense(4, input_dim=8, activation='relu'),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(optimizer='RMSprop', loss='binary_crossentropy', metrics=['accuracy'])


# Define model checkpoin callback
checkpoint = ModelCheckpoint('model_diabetes.h5', monitor='accuracy', save_best_only=True, mode='max')

# fit the model
history = model.fit(x_train, y_train, epochs=1000, verbose=1, callbacks=[checkpoint])
```

```python
# find the epoch with the highest accuracy
max_accuracy_epoch = history.history['accuracy'].index(max(history.history['accuracy'])) + 1
max_accuracy = max(history.history['accuracy'])

# Ambil akurasi dan loss pada epoch ke-961
epoch_961_accuracy = history.history['accuracy'][960]  # indeks dimulai dari 0
epoch_961_loss = history.history['loss'][960]  # jika kamu ingin melihat loss juga

print(f'Epoch 961 - Accuracy: {epoch_961_accuracy}, Loss: {epoch_961_loss}')

print(f'Epoch with highest accuracy: {max_accuracy_epoch}')
print(f'Highest accuracy: {max_accuracy}')
```

```python
# Predict the x_test data
y_pred = model.predict(x_test)

# Convert predictions to binary values (0 or 1)
y_pred_binary = (y_pred > 0.5).astype(int)

# Calculate accuracy
accuracy = (y_pred_binary == y_test).mean()

print(f'Accuracy: {accuracy}')
```
