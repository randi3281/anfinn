# Prediksi Harga Rumah dengan CSV

```python
import pandas as pd

df = pd.read_csv('../../Dataset/data_rumah.csv', encoding='utf-8', sep=';')
df.head()

# Ini Milyar ygy Harganya
```

```python
import numpy as np

df_feature = df.drop(['HARGA', 'NAMA RUMAH', 'NO'], axis=1)
print(df_feature.head())

df_label = df.drop(['LT', 'LB', 'KT', 'KM', 'GRS', 'NAMA RUMAH','NO'], axis=1)
print(df_label.head())
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df_feature, df_label, test_size=0.2, random_state=0)
x_train = x_train.to_numpy()
x_test = x_test.to_numpy()
y_train = y_train.to_numpy().reshape(-1, 1)
y_test = y_test.to_numpy().reshape(-1, 1)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import ModelCheckpoint

model = Sequential([
    Dense(10, input_dim=5, activation='relu'),
    Dense(10, activation='relu'),
    Dense(1)
])

model.compile(optimizer='RMSprop', loss='mse')

checkpoint = ModelCheckpoint('prediksi_harga_rumah_csv.h5', monitor='loss', mode='min', save_best_only=True, verbose=0)

print("Modeling dimulai")
history = model.fit(x_train, y_train, epochs=1000, callbacks=[checkpoint], verbose=0)
print("Modeling selesai")
```

```python
prediksinya = model.predict(x_train)

print("Prediksi Harga Rumahnya:")
for i, prediction in enumerate(prediksinya):
    print(f"Rumah {i+1}: Prediksi Harga = Rp. {prediction[0]:.2f}")
```

```python
errors = np.abs(y_train - prediksinya.flatten())
print(f"Error rata-rata: {np.mean(errors):.2f}")
```
