# Prediksi Harga Rumah

## Percobaan Pertama

```python
import numpy as np

# Dataset Contoh (misalnya, luas rumah dan jumlah kamar tidur)
# x_train = [Luas Rumah, Jumlah Kamar Tidur]
# y_train = Harga Rumah (dalam ribuan dolar)

x_train = np.array([[50, 2], [60, 3], [70, 3], [80, 4], [100, 5]], dtype=float)
y_train = np.array([150, 200, 250, 300, 400], dtype=float)
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import ModelCheckpoint as checkpoint

model = Sequential([
    Dense(units=10, input_shape=[2], activation='relu'),
    Dense(units=10, activation='relu'),
    Dense(units=1)
])

model.compile(optimizer='RMSprop', loss='mean_squared_error')

print('Mulai Modeling')
history = model.fit(x_train, y_train, epochs=10000, callbacks=[checkpoint('prediksi_rumah.h5', save_best_only=True, monitor='loss', mode='min', verbose=0)], verbose=0)
print('Selesai Modeling')
```

```python
predictions = model.predict(x_train)
print("Harga Rumah Prediksi:")
for i, prediction in enumerate(predictions):
    print(f"Rumah {i+1}: Prediksi harga = ${prediction[0]:.2f}K")
```

```python
# Menghitung error rata-rata
errors = np.abs(y_train - predictions.flatten())
print(f"Error rata-rata: {np.mean(errors):.2f}K")
```
